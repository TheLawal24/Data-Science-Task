```python
#Further Comparison Due to flunctuation in Model trends
from bs4 import BeautifulSoup
import requests

def get_model(model_type='text-classification', sort='downloads', limit=20): #Functions

  text_classification = []
  text_generation = []
  url = 'https://huggingface.co/models?pipeline_tag='+model_type+'&sort='+sort
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'lxml')
  models = soup.find_all('article', class_='overview-card-wrapper group/repo')
  c = 0
  for model in models:
    c = c+1
    model_name = model.find('h4', class_='text-md truncate font-mono text-black dark:group-hover/repo:text-yellow-500 group-hover/repo:text-indigo-600 text-smd').text

    if model_type == 'text-classification':
      text_classification.append(model_name)
    elif model_type == 'text-generation':
      text_generation.append(model_name)
  if model_type == 'text-classification':
    return text_classification[:limit]
  elif model_type == 'text-generation':
    return text_generation[:limit]


```
